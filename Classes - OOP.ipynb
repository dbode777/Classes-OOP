```python
class Animal():
    """
    DOCSTRING: This class allows you to enter information on animal objects. You can set the common name, the species, the origin,
    the year an animal was discovered, and it current Cartesian coordinates.

    There are getter methods that can retrieve information on a particular animal. The classification should be updated at the instance the animal entry was initialized.
    """
    # Attributes
    def __init__(self, common_name: str, species: str, classification: dict, origin: str, year_discovered: int, coordinates: dict):
        self._common_name = common_name
        self._species = species
        self._classification = classification
        self._origin = origin
        self._year_discovered = year_discovered
        self._coordinates = coordinates

        x = coordinates[0]
        y = coordinates[1]
        z = coordinates[2]

    # Methods
    def get_common_name(self):
        return self._common_name

    def set_common_name(self):
        self._common_name = common_name
    
    def get_species(self):
        return self._species
    
    def set_species(self):
        self._species = species
    
    def get_classification(self):
        return self._classification
    
    def get_origin(self):
        return self._origin
    
    def set_origin(self):
        self._origin = origin
    
    def get_year_discovered(self):
        return self._year_discovered
    
    def set_year_discovered(self):
        self._year_discovered = year_discovered

    def move(self,x,y,z):
        # x, y, and z are the current coordinates of the animal
        while True: 
            movement = str(input("Which direction would you like to move? Enter left, right, up, down, forwards, or backwards: ").lower())
            if not movement.isalpha():
                print("\nPlease make sure to enter one of the six possible directions.")
                continue
            elif movement == 'left':
                x = x + 1
                return {"x":x,"y":y,"z":z}
            elif movement == 'right':
                x = x - 1
                return {"x":x,"y":y,"z":z}
            elif movement == 'up':
                z = z + 1
                return {"x":x,"y":y,"z":z}
            elif movement == 'down':
                z = z- 1
                return {"x":x,"y":y,"z":z}
            elif movement == 'forwards':
                y = y + 1
                return {"x":x,"y":y,"z":z}
            elif movement == 'backwards':
                y = y - 1
                return {"x":x,"y":y,"z":z}
    
    def metabolize(self):
        print("Munch, munch, munch. Now I'm full. Happy animal I am!")
    
    def hydrate(self):
        print("Slurp, slurp, slurp. Now I'm quenched. Happy animal I am!")
```

```python
class Book():
    """
    DOCSTRING: This is the book class. You can create a book object that has a title, a list of authors, the name of the publisher, the publishing year,
    a description of the book, the number of pages in the book, and a dictionary of the page text. 
    For the dictionary of page text, the keys are the page numbers and the values are the page text.
    """
    def __init__(self, title: str, author_s_: list, publisher: str,
     publishing_year: int, description: str, number_pages: int, page_text: dict):
        self._title = title
        self._authors = author_s_
        self._publisher = publisher
        self._publishing_year = publishing_year
        self._description = description
        self._number_pages = number_pages
        self._page_text = page_text
    
    def get_title(self):
        return self._title

    def set_title(self):
        self._title = title

    def get_authors(self):
        return self._authors
    
    def set_authors(self):
        self._authors = author_s_

    def get_description(self):
        return self._description

    def set_description(self):
        self._description = description

    def get_number_pages(self):
        return self._number_pages

    def set_number_pages(self):
        self._number_pages = number_pages

    def get_publisher(self):
        return self._publisher

    def set_publisher(self):
        self._publisher = publisher

    def get_publishing_year(self):
        return self._publishing_year

    def set_publishing_year(self):
        self._publishing_year = publishing_year
    
    def read_cover(self):
        print(f"{'='*(len(self._title)+6)}\n{' '*3}{self._title}{' '*3}\n{self._authors}\n{'='*(len(self._title)+6)}\n{self._description}")
    
    def read_page(self,page_number = str(1)):
        print(f"Current page: {page_number}")
        print(f"\n{self._page_text[page_number]}")
```

```python
class Vehicle():
    """
    DOCSTRING: The vehicle class can create a vehicle object. The vehicle object can have the following properties defined:
    make, model, color, year, owner, fuel_type, mileage_per_unit_volume, vehicle_type

    With these, a user can retrieve basic information on the vehicle. The methods this object has include the ability to start-up and turn off,
    accelerate or brake, steer, check to see if you have a seatbelt on, and indicate any warnings to the user.
    """

    # Attributes
    def __init__(self, make: str, model: str, color: str, year: int, owner: str,
    fuel_type: str, mileage_per_unit_volume: float, vehicle_type: str):
        self._make = make
        self._model = model
        self._color = color
        self._year = year
        self._owner = owner
        self._fuel_type = fuel_type
        self._avg_mileage = mileage_per_unit_volume
        self._vehicle_type = vehicle_type

    def get_make(self):
        return self._make

    def set_make(self):
        self._make = make

    def get_model(self):
        return self._model

    def set_model(self):
        self._model = model

    def get_color(self):
        return self._color

    def set_color(self):
        self._color = color

    def get_year(self):
        return self._year

    def set_title(self):
        self._year = year

    def get_owner(self):
        return self._owner

    def set_owner(self):
        self._owner = owner

    def get_fuel_type(self):
        return self._fuel_type

    def set_fuel_type(self):
        self._fuel_type = fuel_type

    def get_avg_mileage(self):
        return self._avg_mileage

    def set_avg_mileage(self):
        self._avg_mileage = mileage_per_unit_volume

    def get_vehicle_type(self):
        return self._vehicle_type

    def set_vehicle_type(self):
        self._vehicle_type = vehicle_type

    # Methods
    def turn_on_or_off(self,running = False):
        # Default is to be off
        if running:
            return True
        else:
            return False

    def accelerate(self):
        print("vroom")

    def brake(self):
        print("skrrr")

    def steer(self):
        print(f"The wheels on the {year}, {color} {make} {model} go left and right.")

    def seatbelt_on(self,click = "nay"):
        if click == "nay":
            print("Time for the eternal hell of dinga-ling")
        else:
            print("Good person")

    def warnings(self):
        print("The unending iterations of entropic deterioration may yeet me from this mortal coil.\n Translation: Take me to a mechanic.")
```
